```python
def outlierdetection(data):
    sorted(data)
    
    Q1,Q3=np.percentile(data,[25,75])
    
    IQR=Q3-Q1
    
    lower_range=Q1-(1.5*IQR)
    
    Upper_range=Q3+(1.5*IQR)
    
    return lower_range,Upper_range
```
